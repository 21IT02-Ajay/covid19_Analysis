Top 10 Most Affected Countries 🌍

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Sorted by Ascending Order.
top_country=df[['Country/Region','Confirmed']].sort_values(by='Confirmed',ascending=False).head(10)

#Barplot
sns.barplot(x='Confirmed',y='Country/Region',data=top_country,palette='Reds_r')
plt.title("Top 10 Most Affected Countries")
plt.xlabel("Confirmed")
plt.ylabel("Country/Region")
plt.show()
```

Deaths (vs) Recovered Rate.

```python
#Scatterplot
sns.scatterplot(x=df['Deaths / 100 Cases'],y=df['Recovered / 100 Cases'],data=df)
plt.xlabel('Deaths Rate')
plt.ylabel('Recovered Rate')
plt.title('Death Rate VS Recovered Rate')
plt.grid(True)
plt.show()
```

Correlation Analysis.

```python
#Only For Numbers.
number_df=df.select_dtypes(include='number')

#Heatmap.
sns.heatmap(number_df.corr(),cmap='coolwarm',fmt='.1f',annot=True,linecolor="gray",linewidth=0.5)
```

Deaths (vs) Recovered by country.

```python
#Lineplot.
sns.lineplot(x=df['Deaths'],y=df['Recovered'],data=df)
plt.grid(True)
plt.show()
```

Weekly % Increase in Cases Using Scatterplot.

```python
top_growth=df[['Country/Region','1 week % increase']].sort_values(by='1 week % increase',ascending=False).head(10)

#Scatterplot.
sns.scatterplot(x='1 week % increase',y='Country/Region',data=top_growth,color='g')
plt.grid(True)
plt.show()
```

Weekly % Increase in Cases Using Barplot.

```python
#Barplot.
sns.barplot(x='1 week % increase',y='Country/Region',data=top_growth,palette='Blues_r')
```

Recovery Rate Analysis.

```python
#Histplot.
sns.histplot(df['Recovered / 100 Cases'], bins=20, kde=True,color='g')
plt.xlabel("Recovery Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Recovery Rates Across Countries")
plt.show()
```

Mortality Rate Across WHO Regions.

```python
#Boxplot.
sns.boxplot(x=df['WHO Region'], y=df['Deaths / 100 Cases'], palette='coolwarm')
plt.xlabel("WHO Region")
plt.ylabel("Death Rate")
plt.title("Mortality Rate Across WHO Regions")
plt.xticks(rotation=80)
plt.show()
```

Outliers Remove Using log1p Method.

```python
import numpy as np
log=np.log1p(df['Deaths / 100 Cases'])
print(log)
```

```python
#Boxplot.
sns.boxplot(x=df['WHO Region'], y=log, palette='coolwarm')
plt.xlabel("WHO Region")
plt.ylabel("Death Rate")
plt.title("Mortality Rate Across WHO Regions")
plt.xticks(rotation=80)
plt.show()
```

COVID-19: Confirmed vs Deaths vs Recovered (Log Transformed) Using 3D Plot.

```python
from mpl_toolkits.mplot3d import Axes3D


# Apply log1p transformation to handle outliers.
df["log_confirmed"] = np.log1p(df["Confirmed"])
df["log_deaths"] = np.log1p(df["Deaths"])
df["log_recovered"] = np.log1p(df["Recovered"])

#3d Plot.
fig = plt.figure(figsize=(10, 7))
axes=fig.add_subplot(111,projection='3d')

#Scatterplot.
diagram=axes.scatter(df['log_confirmed'],df['log_deaths'],df['log_recovered'],c=df['Active'],cmap='coolwarm')
axes.set_title('COVID-19: Confirmed vs Deaths vs Recovered (Log Transformed)')
axes.set_xlabel('log Confirmed Cases')
axes.set_ylabel('log Deaths Cases')
axes.set_zlabel('log Recovered Cases')


#colorbar.
plt.colorbar(diagram,label='Active Cases')
plt.show()
```

```python
import plotly.express as px

# Apply log1p transformation to handle outliers.
df["log_confirmed"] = np.log1p(df["Confirmed"])
df["log_deaths"] = np.log1p(df["Deaths"])
df["log_recovered"] = np.log1p(df["Recovered"])

# Create interactive 3D scatter plot
fig = px.scatter_3d( x=df["log_confirmed"],y=df["log_deaths"],z=df["log_recovered"],color=df["Active"],hover_name=df["Country/Region"],
                     title="COVID-19: 3D Interactive Plot (Zoom & Pan)")

# Show figure
fig.show()
```
